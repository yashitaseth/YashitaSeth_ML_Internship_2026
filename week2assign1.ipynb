```python
#IMPORT LIBRARIES
import numpy as np #for numerical computation
import pandas as pd #for reading and manipulating datasets
#ML Agorithms
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
#to split dataset into training and testing models
from sklearn.model_selection import train_test_split
#to evaluate classification models
from sklearn.metrics import accuracy_score, confusion_matrix
#to convert text columns into numerical values
from sklearn.preprocessing import LabelEncoder

print("# Part A: Understanding the dataset")
print("#1 Load the dataset and display the first five records.")
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Dataset 2.csv")
print(df.head(5))
print("#2 Determine the number of rows and columns in the dataset.")
rows, cols = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {cols}")
print("#3 Display all column names.")
print("Column names:")
print(list(df.columns))
print("#4 Identify numerical and categorical features.")
numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = df.select_dtypes(exclude=[np.number]).columns.tolist()
print("Numerical features:" ,numerical_features)
print("Categorical features:",categorical_features)
print("#5 Check whether the dataset contains missing values.")
print("Missing values per column:")
print(df.isnull().sum())

print("# Part B: Exploratory Data Analysis")
print("#6 Calculate the average age of users.")
avg_age = df['Age'].mean()
print("Average age of users:", avg_age ,"years")
print("#7 Determine the average watch hours per week.")
avg_watch_hours = df['WatchHoursPerWeek'].mean()
print("Average watch hours per week",avg_watch_hours, "hours")
print("#8 Find the average monthly spending of users.")
avg_spend = df['MonthlySpend'].mean()
print("Average monthly spending: $" , avg_spend)
print("#9 Count the number of users in each subscription category.")
print("User counts per subscription category:")
print(df['SubscriptionType'].value_counts())
print("#10 Determine the percentage of users who renewed their subscriptions.")
renewal_pct = (df['SubscriptionRenewed'].value_counts(normalize=True).get('Yes', 0)) * 100
print("Percentage of users who renewed: ", renewal_pct ,"%")

print("# Part C: Data Preparation")
print("#11 Convert categorical features into numerical form.")
df_new = df.copy()
df_new = df_new.drop(columns=['UserID'])
df_new = pd.get_dummies(df_new, columns=['Gender', 'SubscriptionType', 'FavoriteGenre'], drop_first=True)
df_new['SubscriptionRenewed'] = df_new['SubscriptionRenewed'].map({'Yes': 1, 'No': 0})
print("New dataset with converted categorical features:")
print(df_new.head())
print("#12 Define the feature set (X) and target variable (y) for subscription renewal prediction.")
X_class = df_new.drop(columns=['SubscriptionRenewed'])
y_class = df_new['SubscriptionRenewed']
print("#13 Split the dataset into training and testing sets.")
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_class, y_class, test_size=0.2, random_state=42)
print(f"Data split. Training set shape: {X_train_c.shape}, Testing set shape: {X_test_c.shape}")

print("# Part D: Decision Tree Classification")
print("#14 Train a Decision Tree model to predict whether a user will renew their subscription.")
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_c, y_train_c)
print("Decision Tree model to predict whether a user will renew their subscription has been successfully trained")
print("#15 Evaluate the model using accuracy.")
y_pred_dt = dt_model.fit(X_train_c, y_train_c).predict(X_test_c)
dt_accuracy = accuracy_score(y_test_c, y_pred_dt)
print("Decision Tree Model Accuracy:" ,dt_accuracy * 100,"%")
print("#16 Generate and interpret the confusion matrix.")
cm_dt = confusion_matrix(y_test_c, y_pred_dt)
print("Decision Tree Confusion Matrix:")
print(cm_dt)
print(f"Interpretation: \n"
      f" - True Negatives (Correctly predicted non-renewals): {cm_dt[0][0]}\n"
      f" - False Positives (Incorrectly predicted renewals): {cm_dt[0][1]}\n"
      f" - False Negatives (Incorrectly predicted non-renewals): {cm_dt[1][0]}\n"
      f" - True Positives (Correctly predicted renewals): {cm_dt[1][1]}")

print("# Part E: K-Nearest Neighbors (KNN)")
print("#17 Train a KNN classifier with K = 5.")
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_c, y_train_c)
print("KNN model with K=5 successfully trained.")
print("#18 Compare the accuracy of KNN with the Decision Tree model.")
y_pred_knn = knn_model.predict(X_test_c)
knn_accuracy = accuracy_score(y_test_c, y_pred_knn)
print("KNN Model Accuracy: ",knn_accuracy * 100,"%")
print("Decision Tree Model Accuracy:",dt_accuracy * 100,"%")
if dt_accuracy > knn_accuracy:
    print("Decision Tree model is more accurate.")
elif knn_accuracy > dt_accuracy:
    print("KNN model is more accurate.")
else:
    print("CBoth models are equally accurate.")
print("Part F: Linear Regression")
print("#19 Train a Linear Regression model to predict monthly spending.")
X_reg = df_new.drop(columns=['MonthlySpend'])
y_reg = df_new['MonthlySpend']
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
lr_model = LinearRegression()
lr_model.fit(X_train_r, y_train_r)
print("Linear Regression model to predict MonthlySpend trained successfully.")
print("#20 Predict the monthly spending for a new user and interpret the result.")
sample_user = X_test_r.iloc[[0]]
predicted_spend = lr_model.predict(sample_user)[0]
actual_spend = y_test_r.iloc[0]
print("Predicted Monthly Spend for the sample user: $", predicted_spend)
print("Actual Monthly Spend for this sample user: $", actual_spend)
```
